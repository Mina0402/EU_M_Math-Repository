```python
import numpy as np
import numpy.random as random
import scipy as sp
from pandas import Series, DataFrame
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

import sklearn

%precision 3
```

```python
#重回帰
import requests, zipfile
import io

url='https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
res=requests.get(url).content

auto=pd.read_csv(io.StringIO(res.decode('utf-8')), header=None)

auto.columns=['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration','num-of-doors',
             'body-style','drive-wheels','engine-location','wheel-base','length','width','height',
             'curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore',
             'stroke','compression-ration','horsepower','peak-rpm','city-mpg','highway-mpg','price']
```

```python
print('自動車データの形式:',format(auto.shape))
```

```python
auto.head()
```

```python
#データ整理
auto=auto[['price', 'horsepower', 'width', 'height']]
auto.isin(['?']).sum()
```

```python
auto=auto.replace('?', np.nan).dropna()
print('自動車データ形式:',format(auto.shape))
```

```python
#型の変換
print('データ型の確認(型変換前)\n', format(auto.dtypes))
```

```python
auto=auto.assign(price=pd.to_numeric(auto.price))
auto=auto.assign(horsepower=pd.to_numeric(auto.horsepower))
print('データ型の確認(型変換後)\n', format(auto.dtypes))
```

```python
#相関の確認
auto.corr()
```

```python
#モデル構築と評価
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X=auto.drop('price', axis=1)
y=auto['price']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=0)

#重回帰クラスの初期化と学習
model=LinearRegression()
model.fit(X_train,y_train)

#決定係数の表示
print('決定係数(train):{:.3f}'.format(model.score(X_train,y_train)))
print('決定係数(test):{:.3f}'.format(model.score(X_test,y_test)))

#回帰係数と切片の表示
print('\n回帰係数\n():',format(pd.Series(model.coef_, index=X.columns)))
print('切片:{:.3f}',format(model.intercept_))
```

```python
#ロジスティクス回帰
url2='https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
res2=requests.get(url2).content

adult=pd.read_csv(io.StringIO(res2.decode('utf-8')), header=None)

adult.columns=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
               'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss', 'hours-per-week',
               'native-country', 'flg-50K']

print('データの形式:()'.format(adult.shape))
print('欠損の数:()',format(adult.isnull().sum().sum()))

adult.head()
```

```python
adult.groupby('flg-50K').size()
```

```python
#「fin_flg」カラムを追加し、「flg-50K」カラムが「>50K」なら1、そうでなければ0で
adult['fin_flg']=adult['flg-50K'].map(lambda x: 1 if x==' >50K' else 0)
adult.groupby('fin_flg').size()
```

```python
#モデル構築と評価
from sklearn.linear_model import LogisticRegression

X2=adult[['age','fnlwgt','education-num','capital-gain','capital-loss']]
y2=adult['fin_flg']

X2_train, X2_test, y2_train, y2_test=train_test_split(X2, y2, test_size=0.5, random_state=0)

model2=LogisticRegression()
model2.fit(X2_train,y2_train)

print('正解率(train):{:.3f}'.format(model2.score(X2_train, y2_train)))
print('正解率(test):{:.3f}'.format(model2.score(X2_test,y2_test)))
```

```python
#予測精度を上げるスケーリング
from sklearn.preprocessing import StandardScaler

#標準化処理
sc=StandardScaler()
sc.fit(X2_train)
X2_train_std=sc.transform(X2_train)
X2_test_std=sc.transform(X2_test)

model2=LogisticRegression()
model2.fit(X2_train_std,y2_train)

print('正解率(train):{:.3f}'.format(model2.score(X2_train_std, y2_train)))
print('正解率(test):{:.3f}'.format(model2.score(X2_test_std,y2_test)))
```

```python
import numpy as np
import numpy.random as random
import scipy as sp
from pandas import Series, DataFrame
import pandas as pd

import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
%matplotlib inline

import sklearn

%precision 3
```

```python
#重回帰
import requests, zipfile
import io

url='https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data'
res=requests.get(url).content

auto=pd.read_csv(io.StringIO(res.decode('utf-8')), header=None)

auto.columns=['symboling', 'normalized-losses', 'make', 'fuel-type', 'aspiration','num-of-doors',
             'body-style','drive-wheels','engine-location','wheel-base','length','width','height',
             'curb-weight','engine-type','num-of-cylinders','engine-size','fuel-system','bore',
             'stroke','compression-ration','horsepower','peak-rpm','city-mpg','highway-mpg','price']
```

```python
print('自動車データの形式:',format(auto.shape))
```

```python
auto.head()
```

```python
#データ整理
auto=auto[['price', 'horsepower', 'width', 'height']]
auto.isin(['?']).sum()
```

```python
auto=auto.replace('?', np.nan).dropna()
print('自動車データ形式:',format(auto.shape))
```

```python
#型の変換
print('データ型の確認(型変換前)\n', format(auto.dtypes))
```

```python
auto=auto.assign(price=pd.to_numeric(auto.price))
auto=auto.assign(horsepower=pd.to_numeric(auto.horsepower))
print('データ型の確認(型変換後)\n', format(auto.dtypes))
```

```python
#相関の確認
auto.corr()
```

```python
#モデル構築と評価
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X=auto.drop('price', axis=1)
y=auto['price']

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.5,random_state=0)

#重回帰クラスの初期化と学習
model=LinearRegression()
model.fit(X_train,y_train)

#決定係数の表示
print('決定係数(train):{:.3f}'.format(model.score(X_train,y_train)))
print('決定係数(test):{:.3f}'.format(model.score(X_test,y_test)))

#回帰係数と切片の表示
print('\n回帰係数\n():',format(pd.Series(model.coef_, index=X.columns)))
print('切片:{:.3f}',format(model.intercept_))
```

```python
#リッジ回帰用のクラス
from sklearn.linear_model import Ridge

linear=LinearRegression()
ridge=Ridge(random_state=0)

for model in [linear, ridge]:
  model.fit(X_train,y_train)
  print('{}(train):{:.6f}'.format(model.__class__.__name__, model.score(X_train, y_train)))
  print('{}(test):{:.6f}'.format(model.__class__.__name__, model.score(X_test, y_test)))
```

```python
#練習問題8-4
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error, r2_score

scaler = StandardScaler()
X4_train_scaled = scaler.fit_transform(X1_train)
X4_test_scaled = scaler.transform(X1_test)

# ラッソ回帰モデルの初期化と学習
lasso = Lasso(alpha=1.0, random_state=0)
lasso.fit(X4_train_scaled, y1_train)

# 訓練データとテストデータに対する予測
y4_train_pred = lasso.predict(X4_train_scaled)
y4_test_pred = lasso.predict(X4_test_scaled)

# モデルの評価
print('訓練データに対する決定係数: {:.6f}'.format(r2_score(y1_train, y4_train_pred)))
print('テストデータに対する決定係数: {:.6f}'.format(r2_score(y1_test, y4_test_pred)))
print('訓練データに対するRMSE: {:.6f}'.format(mean_squared_error(y1_train, y4_train_pred, squared=False)))
print('テストデータに対するRMSE: {:.6f}'.format(mean_squared_error(y1_test, y4_test_pred, squared=False)))

# 回帰係数の表示
print('\n回帰係数:')
print(pd.Series(lasso.coef_, index=X1.columns))
print('切片: {:.6f}'.format(lasso.intercept_))
```
